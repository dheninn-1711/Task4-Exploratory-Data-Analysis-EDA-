```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic train dataset
Titanic_train_df = pd.read_csv("C:/Users/dheni/OneDrive/Documents/Titanic_train.csv")
Titanic_train_df
```

```python
# Load the Titanic test dataset
Titanic_test_df = pd.read_csv("C:/Users/dheni/OneDrive/Documents/Titanic_test.csv")
Titanic_test_df
```

```python
# Datasets Info
print("Titanic_train Info: ")
Titanic_train_df.info()
print("Titanic_test Info: ")
Titanic_test_df.info()
```

```python
# Drop the cabin column
Titanic_train_df.drop(['Cabin'],axis=1,inplace=True)
Titanic_test_df.drop(['Cabin'],axis=1,inplace=True)
```

```python
# Basic filling the null values in Age column
Titanic_train_df['Age'].fillna(Titanic_train_df['Age'].median(), inplace=True)
Titanic_test_df['Age'].fillna(Titanic_test_df['Age'].median(), inplace=True)
```

```python
# Basic filling the null values in Embarked column
Titanic_train_df['Embarked'].fillna(Titanic_train_df['Embarked'].mode()[0], inplace=True)
```

```python
# Basic filling the null values in Fare column
Titanic_test_df['Fare'].fillna(Titanic_test_df['Fare'].median(), inplace=True)
```

```python
# Checking for missing values
print("Titanic_train_df: ")
print(Titanic_train_df.isnull().sum())
print("Titanic_test_df: ")
print(Titanic_test_df.isnull().sum())
```

```python
# Describe Titanic_test
Titanic_test_df.describe()
```

```python
# Unique values
print("Unique Values in Train: ") 
print(Titanic_train_df.nunique())
print("Unique Values in Test: ") 
print(Titanic_test_df.nunique())
```

```python
# Value_counts
print("Survival Counts: ") 
print(Titanic_train_df['Survived'].value_counts())
print("Sex Counts: ") 
print(Titanic_train_df['Sex'].value_counts())
print("Pclass Counts: ")
print(Titanic_train_df['Pclass'].value_counts())
print("Embarked Counts: ") 
print(Titanic_train_df['Embarked'].value_counts())
```

```python
# Histograms
Titanic_train_df[['Age', 'Fare']].hist(bins=20, figsize=(10, 5))
plt.suptitle("Histograms: Age and Fare (Train Data)")
plt.show()
```

- The Age distribution is right-skewed, with a strong concentration of passengers in their 20s.
- The Fare distribution is heavily right-skewed, with most passengers paying low fares and a few paying extremely high amounts.

```python
# Count Plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.countplot(x='Sex', data=Titanic_train_df, ax=axes[0])
sns.countplot(x='Pclass', data=Titanic_train_df, ax=axes[1])
sns.countplot(x='Embarked', data=Titanic_train_df, ax=axes[2])
fig.suptitle("Count Plots for Categorical Features (Train Data)")
plt.show()
```

- The dataset has significantly more male passengers than female passengers.
- Most passengers traveled in 3rd class, while 1st and 2nd classes had fewer travelers.
- A majority of passengers embarked from port 'S', with fewer from 'C' and even fewer from 'Q'.

```python
# Boxplot
sns.boxplot(x='Survived', y='Age', data=Titanic_train_df)
plt.title("Survival by Age")
plt.show()
```

- In both the "Survived" (1) and "Not Survived" (0) groups, there are several outliers — especially above the typical range (ages 50–80).
- Both groups show a wide age range with several outliers, especially among older ages.
- Among survivors, there are fewer high-age outliers compared to non-survivors, suggesting older individuals had lower survival chances overall.
- This pattern reflects potential age-based survival biases, likely influenced by evacuation priorities (e.g., "women and children first").

```python
# Boxplot
sns.boxplot(x='Survived', y='Fare', data=Titanic_train_df)
plt.title("Survival by Fare")
plt.show()
```

- Survivors generally paid higher fares than non-survivors, as shown by the higher median and wider spread.
- High-fare outliers (above 100 and even up to 500) are much more common among survivors, suggesting that wealthier passengers had better survival chances.
- Non-survivors mostly clustered around lower fare values, with fewer high-fare outliers.
- This reflects a strong link between fare (likely class or cabin quality) and survival probability.

```python
# Converting datatype to string
Titanic_train_df['Survived'] = Titanic_train_df['Survived'].astype(str)
```

```python
# Countplot for survival by categorical features
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.countplot(x='Sex', hue='Survived', data=Titanic_train_df, ax=axes[0])
sns.countplot(x='Pclass', hue='Survived', data=Titanic_train_df, ax=axes[1])
sns.countplot(x='Embarked', hue='Survived', data=Titanic_train_df, ax=axes[2])
fig.suptitle("Survival by Categorical Features")
plt.show()
```

- Females had a much higher survival rate compared to males, highlighting a strong gender-based survival advantage.
- Passengers from 1st class had the highest survival rate, while 3rd class passengers had the lowest, showing a clear link between class and survival.
- Those who embarked from 'C' had a better survival rate compared to passengers from 'S' and 'Q'.
- Overall, socioeconomic factors class, port of embarkation and gender played major roles in survival outcomes.

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(Titanic_train_df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Train Data)")
plt.show()
```

- Fare and Pclass show a strong negative correlation: passengers in higher classes (lower Pclass number) paid higher fares, matching earlier plots where 1st class had better survival rates.
- SibSp and Parch (family relations) have a positive correlation, meaning passengers traveling with siblings often also traveled with parents/children, possibly influencing survival (families may have been prioritized).
- Age has weak correlations with other features, indicating age varied widely across classes and fare prices — aligning with the broad age distributions seen earlier.
- Fare shows a mild positive relationship with family size (SibSp, Parch), suggesting families might have paid more overall, but the connection isn't strong individually.
- Overall, Pclass and Fare emerge as the strongest predictors among the available numeric features, confirming trends observed in survival by class and fare.

```python
# Pairplot
sns.pairplot(Titanic_train_df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

- Survivors (orange) are more concentrated in lower Pclass (higher class) compared to non-survivors (blue).
- Age distribution shows that younger passengers had a slightly better survival rate.
- Higher fares are associated with better survival chances, reflecting that higher-paying passengers were more likely in safer classes.

```python
# Histplot for Age Distribution
sns.histplot(Titanic_train_df['Age'], label='Train', color='blue', kde=True)
sns.histplot(Titanic_test_df['Age'], label='Test', color='orange', kde=True)
plt.title("Age Distribution: Train vs Test")
plt.legend()
plt.show()
```

- Both datasets follow a similar overall distribution, with a peak around ages 20–30.
- The training set (blue) has a strong spike around age 29–30, likely due to imputation or data clustering.
- The test set (orange) is smoother and slightly more evenly spread, but overall trends remain comparable.

```python
# Histplot for Fare Distribution
sns.histplot(Titanic_train_df['Fare'], label='Train', color='green', kde=True)
sns.histplot(Titanic_test_df['Fare'], label='Test', color='red', kde=True)
plt.title("Fare Distribution: Train vs Test")
plt.legend()
plt.show()
```

- Both distributions are heavily right-skewed, with most passengers paying low fares (under $50).

- There are a few outliers with fares exceeding $200, but they are rare.

### 🔍 Summary of Findings:
- **Gender**: Females had significantly higher survival rates than males, likely due to the "women and children first" evacuation policy.

- **Class (Pclass)**: Passengers in 1st class had a much higher survival rate compared to those in 2nd and 3rd class, reflecting the socio-economic disparities in access to lifeboats and crew assistance.

- **Age**: Children under approximately 15 years of age had higher survival rates. However, missing age data must be addressed to ensure accurate analysis.

- **Fare**: A higher fare is positively correlated with survival, indicating that wealthier passengers had a better chance of survival.

- **Embarked**: There is a slight variation in survival rates based on the port of embarkation — passengers who embarked from Cherbourg (C) had the highest survival rates, followed by (Q), and (S).

- **Correlation Summary**:

   - Pclass is negatively correlated with Survived (lower class → lower survival).

   - Fare is positively correlated with Survived (higher fare → higher survival).

- **Missing Data**:

   - Age: Contains missing values that need to be imputed using appropriate statistical or machine learning techniques.

   - Cabin: A large portion of cabin data is missing, which may affect detailed analysis of accommodation location or ship section.
- **Test Dataset**: The test dataset exhibits a similar distribution to the training dataset, making it suitable for validating models without introducing significant bias.
